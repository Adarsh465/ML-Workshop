# Face Detection

```
!pip install opencv-python
```

```
import cv2
```

```
! wget "https://image.shutterstock.com/image-photo/photo-happy-group-friends-sitting-260nw-1071557390.jpg"
```

```
group_photo = plt.imread('/content/group_photo.jpg')
plt.imshow(group_photo)
```

```
! wget "https://www.dropbox.com/s/5o64ru428jlb29g/haarcascade_frontalface_alt.xml"
```

```
face_model = cv2.CascadeClassifier("/content/haarcascade_frontalface_alt.xml")
```

```
faces = face_model.detectMultiScale(group_photo)
```

```
faces
```

```
for face in faces:
  x,y,w,h = face
  group_photo = cv2.rectangle(group_photo, (x,y) , (x+w, y+h), (255, 0, 0))
```

```
plt.imshow(group_photo)
```
